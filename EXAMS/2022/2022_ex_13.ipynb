```python
import numpy as np
```

```python
C = np.loadtxt('data/cost.txt', dtype=int)
```

```python
# Calculate s0 (delta x = 0 constant)

s0 = C.sum(axis=1) # sum each row 
s0 = np.argmin(s0) # find the index of the minimum sum
print(f's0 = {s0}')
```

```python
# Step 3: calculate s5 (delta x = 5 constant)

s5 = np.argmin(C, axis=0)  # Find best row for each column
print(f's5 = {s5}')
```

```python
d = abs(s0 - s5)  # s0 is scalar 1, s5 is array
print(f'Question 13: {d.mean()}')
```
